```python
# Que.1: Greeting user by name

# Ask the user to input their name
name = input("Hello, what is your name? ")

# Display a greeting message with name 
print("Hello,", name + ". Good to meet you!")
```

```python
# Que.2: Convert Celsius to Fahrenheit

# Ask the user to enter temperature in Celsius
celsius = float(input("Enter a temperature in Celsius: "))

# Conversion of Celsius to Fahrenheit
fahrenheit = (celsius * 9 / 5) + 32

# Display the result
print(celsius, "C is equivalent to", fahrenheit, "F.")
```

```python
# Que.3: Calculate lab groups and leftover students


students = int(input("How many students? "))       # Ask for total number of students

group_size = int(input("Required group size? "))   # Ask for required group size

groups = students // group_size    # Calculate number of full groups


leftover = students % group_size   # Calculate leftover students

# Output with fixed grammer
if leftover == 1:
    print("There will be", groups, "groups with", leftover, "student left over.")
else:
    print("There will be", groups, "groups with", leftover, "students left over.")
```

```python
# Que.4: Distribute sweets among pupils who attended the class


sweets = int(input("How many sweets are there? "))     # Ask the teacher for total sweets

pupils = int(input("How many pupils are present? "))   # Ask for number of pupils

sweets_per_pupil = sweets // pupils    # Calculate sweets per pupil

leftover_sweets = sweets % pupils      # Calculate leftover sweets

# Display results
print("Each pupil will receive", sweets_per_pupil, "sweets.")
print("There will be", leftover_sweets, "sweets left over.")
```
