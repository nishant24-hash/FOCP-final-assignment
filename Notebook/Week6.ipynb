```python
# Que.1: Convert an integer to binary

def to_binary(number):
    """Returns the binary representation of a positive integer"""
    return bin(number)[2:]  # Remove '0b' prefix

# Checks the function
value = int(input("Enter a positive integer: "))
print("Binary:", to_binary(value))
```

```python
# Que.2: Find all factors of an integer

def find_factors(number):
    
    factors = []    #Returns a list of factors of the given integer

    for i in range(1, number + 1):
        if number % i == 0:
            factors.append(i)

    return factors

# Checks the function
num = int(input("Enter an integer: "))
print("Factors:", find_factors(num))
```

```python
# Que.3: Determine whether a number is prime

def is_prime(number):
    # Returns True if number is prime, otherwise False
    if number <= 1:
        return False

    for i in range(2, number):
        if number % i == 0:
            return False

    return True

# Checks the function
num = int(input("Enter an integer: "))
print("Is prime?", is_prime(num))
```

```python
# Que.4: Message encryption program

def simple_encrypt(message):
    # Remove all spaces from the message
    message_no_spaces = message.replace(" ", "")
    
    # Reverse the message
    encrypted_message = message_no_spaces[::-1]
    
    # Return the encrypted message
    return encrypted_message

# Ask the user to type a message
text = input("Enter a message to encrypt: ")

# Show the encrypted message
print("Encrypted message:", simple_encrypt(text))
```

```python
# Que.5: Encrypt message by spacing letters randomly

import random
import string

def hide_message(message):
    """
    Encrypts a message by spreading its characters at random intervals
    Returns encrypted message and the interval used
    """
    interval = random.randint(2, 20)
    encrypted = ""

    for char in message.replace(" ", ""):
        encrypted += char
        for _ in range(interval - 1):
            encrypted += random.choice(string.ascii_lowercase)

    return encrypted, interval

# Checks the function
msg = input("Enter a message: ")
encrypted_msg, used_interval = hide_message(msg)

print("Encrypted message:", encrypted_msg)
print("Interval used:", used_interval)
```

```python
# Que.6: Decrypt a message hidden at fixed intervals

def decrypt_message(encrypted, interval):
    
    decrypted = ""        # Start with an empty string to store the decrypted message
    
    # Go through the encrypted message, skipping 'interval' characters each time
    for i in range(0, len(encrypted), interval):
        decrypted += encrypted[i]  # Add the character to the decrypted message
    
    return decrypted  # Return the final decrypted message

# Ask the user for the encrypted message
encrypted_text = input("Enter the encrypted message: ")

# Ask the user for the interval number
interval = int(input("Enter the interval used to encrypt: "))

# Show the decrypted message
print("Decrypted message:", decrypt_message(encrypted_text, interval))
```
