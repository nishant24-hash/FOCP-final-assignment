```python
# Function to return sorted list of unique letters in a string

def unique_letters(text):
    
    #Takes a string and returns a sorted list of unique letters.
    
    letters_set = set(text)           # Convert the string to a set to remove duplicates
    letters_list = list(letters_set)  # Convert the set back to a list
    letters_list.sort()               # Sort the list alphabetically
    return letters_list               # Return the sorted list

# Ask the user to enter a string
user_input = input("Enter a string: ")

# Call the function and store the result
result = unique_letters(user_input)

# Show the sorted unique letters
print("Sorted unique letters:", result)
```

```python
# Que.2: Set operations on letters in words

def letters_in_either(word1, word2):
    #Letters that appear in at least one of the words
    return sorted(set(word1.lower()) | set(word2.lower()))

def letters_in_both(word1, word2):
    #Letters that appear in both words
    return sorted(set(word1.lower()) & set(word2.lower()))

def letters_in_one_only(word1, word2):
    #Letters that appear in either word, but not both
    return sorted(set(word1.lower()) ^ set(word2.lower()))

# Check the functions
print(letters_in_either("cheese", "bread"))
print(letters_in_both("cheese", "bread"))
print(letters_in_one_only("cheese", "bread"))
```

```python
# Que.3: Country and capital management system

capitals = {}  # Dictionary to store country-capital pairs

while True:
    country = input("Enter a country name (or press Enter to quit): ")

    # Exit condition
    if country == "":
        print("Program ended.")
        break

    # Normalize input (i.e. case-insensitive)
    country_key = country.strip().lower()

    if country_key in capitals:
        print("The capital of", country.title(), "is", capitals[country_key])
    else:
        capital = input("I don't know the capital. Please enter it: ")
        capitals[country_key] = capital.strip().title()
        print("Thank you, for the information.")
```

```python
# Que.4: Frequency analysis of letters in a message

def frequency_analysis(message):
    
    # Reports the six most common letters (case-insensitive)  and how often they appear
    
    message = message.lower()

    # Initialise dictionary with zero counts
    counts = {}

    for char in message:
        if char.isalpha():
            if char not in counts:
                counts[char] = 0
            counts[char] += 1

    # Sort dictionary items by frequency (descending)
    sorted_letters = sorted(counts.items(), key=lambda item: item[1], reverse=True)

    # Display the six most common letters
    print("Six most common letters:")
    for letter, count in sorted_letters[:6]:
        print(letter, ":", count)

# Check the program
text = input("Enter encrypted message: ")
frequency_analysis(text)
```
