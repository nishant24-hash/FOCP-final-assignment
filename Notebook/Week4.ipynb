```python
"""Que.1: Check if a number is between 0 and 100"""

def is_valid_range(number):
    # Check if the number is within the range
    if 0 <= number <= 100:
        return True
    else:
        return False

# Ask the user to enter a number
number = int(input("Enter a number: "))

# Displays the result
print(is_valid_range(number))

```

```python
"""Que.2: Count uppercase and lowercase letters in a string"""

def count_case(text):
    # Returns number of uppercase and lowercase letters"""
    uppercase = 0  # Stores the count of uppercase letters
    lowercase = 0  # Stores the count of lowercase letters

    for char in text:
        if char.isupper():    # Check if the character is uppercase
            uppercase += 1
        elif char.islower():  # Check if the character is uppercase
            lowercase += 1

    return uppercase, lowercase

sentence = input("Enter a string: ")    # Ask the user to enter a sentence or word
upper, lower = count_case(sentence)

print("Uppercase letters:", upper)      # Displays the result
print("Lowercase letters:", lower)
```

```python
""" Que.3: Greeting with proper capitalization, ie. first character of the string is always capital"""

name = input("Enter your name: ")

if name == "":
    print("Hello, Stranger!")
else:
    # Capitalize first letter and make the rest lowercase
    formatted_name = name.capitalize()
    print("Hello,", formatted_name + ". Good to meet you!")
```

```python
""" Que.4: Remove the last character from a string"""

def remove_last_char(text):
    #Removes last character if length > 1, otherwise returns unchanged
    if len(text) <= 1:
        return text
    return text[:-1]

# Checks the function
user_text = input("Enter a string: ")
print("Result:", remove_last_char(user_text))
```

```python
"""Que.5: Celsius to Fahrenheit and vice-versa conversion functions"""

def celsius_to_fahrenheit(c):
    #Converts Celsius to Fahrenheit
    return (c * 9 / 5) + 32

def fahrenheit_to_celsius(f):
    #Converts Fahrenheit to Celsius
    return (f - 32) * 5 / 9

# Check the functions
c = float(input("Enter temperature in Celsius: "))
print("Fahrenheit:", celsius_to_fahrenheit(c))

f = float(input("Enter temperature in Fahrenheit: "))
print("Celsius:", fahrenheit_to_celsius(f))
```

```python
""" Que.6: Convert Celsius input (like 25C, 65C followed by letter C) to Fahrenheit"""

temperature = input("Enter temperature (e.g. 25C): ")

# Remove the last character (C) and convert to number
celsius = float(temperature[:-1])

fahrenheit = (celsius * 9 / 5) + 32

print(f"{fahrenheit}F")
```

```python
""" Que.7: Read 6 temperatures and display statistics """

import statistics   # Import statistics module to calculate the mean

temperatures = []   # Create an empty list to store temperatures

# Ask the user to enter 6 temperatures
for i in range(6):
    temp_input = input("Enter temperature (e.g. 20C): ")
    temp_value = float(temp_input[:-1])  # Remove 'C'
    temperatures.append(temp_value)
#Display highest, lowest and average temperature
print("Maximum:", max(temperatures), "C")
print("Minimum:", min(temperatures), "C")
print("Mean:", statistics.mean(temperatures), "C")
```

```python
""" Que.8: Read temperatures until Enter is pressed"""

import statistics   # Import statistics module to calculate the mean

temperatures = []   # Create an empty list to store temperatures

while True:
    temp_input = input("Enter temperature (or press Enter to finish): ")

    # Stop when user presses Enter
    if temp_input == "":
        break

    temp_value = float(temp_input[:-1])  # Remove 'C'
    temperatures.append(temp_value)

# Display results if at least one value was entered
if temperatures:
    print("Maximum:", max(temperatures), "C")
    print("Minimum:", min(temperatures), "C")
    print("Mean:", statistics.mean(temperatures), "C")
else:
    print("No temperatures entered.")
```
