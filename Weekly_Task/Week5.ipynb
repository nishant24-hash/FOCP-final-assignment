```python
# Que.1: Display the operating system platform

# Uses the sys module to detect OS information
import sys

print("Operating system platform:", sys.platform)
```

```python
# Que.2: To Count how many command-line arguments were provided

import sys

# sys.argv includes the program name, so subtract 1
argument_count = len(sys.argv) - 1

print("Number of arguments provided:", argument_count)
```

```python
# Que.3: Print the shortest command-line argument

import sys   # Import sys module to work with command-line arguments

# Check if the user has typed any arguments
if len(sys.argv) <= 1:
    print("No arguments were given.")
    sys.exit()   # Stop the program if no arguments are found
    
# Get the arguments (skip the program name)
arguments = sys.argv[1:]

# Take the first argument as the shortest one for now
shortest = arguments[0]

# Check each argument one by one
for arg in arguments:
    # If this argument is shorter, update the shortest value
    if len(arg) < len(shortest):
        shortest = arg

# Show the shortest argument
print("The shortest argument is:", shortest)
```

```python
# Que.4: Check if a website link is working

import sys                 # Used to read command-line input
import urllib.request      # Used to open a website link

# Check if the user has entered a URL
if len(sys.argv) <= 1:
    print("Please enter a website URL.")
    sys.exit()             # Stop the program if no URL is given

# Get the URL from the command line
url = sys.argv[1]

try:
    # Try to open the website
    response = urllib.request.urlopen(url)

    # If successful, show a success message
    print("The website is working.")

    # Show the HTTP status code (for example, 200)
    print("HTTP Status Code:", response.getcode())

except:
    # This runs if the website cannot be opened
    print("The website is not reachable.")
```

```python
# Que.5: Process temperature readings from command-line arguments

import sys
import statistics

# Check if temperatures were provided
if len(sys.argv) <= 1:
    print("Error: No temperature values provided.")
    sys.exit()

temperatures = []

# Convert arguments to floats
for value in sys.argv[1:]:
    temperatures.append(float(value))

# Display max, min and mean
print("Maximum:", max(temperatures))
print("Minimum:", min(temperatures))
print("Mean:", statistics.mean(temperatures))
```

```python
# Que.6: Create a backup copy of a file

import sys
import shutil

# Check if filename is provided
if len(sys.argv) <= 1:
    print("Error: No file name provided.")
    sys.exit()

original_file = sys.argv[1]
backup_file = original_file + ".bak"

try:
    # Copy file contents exactly
    shutil.copy(original_file, backup_file)
    print("Backup created:", backup_file)
except:
    print("Error: File could not be backed up.")
```
