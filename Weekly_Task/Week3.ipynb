```python
# Que.1: Greeting with default for empty input

# Ask the user to input their name
name = input("Hello, what is your name? ")

# Check if the user pressed Enter without typing anything
if name == "":
    print("Hello, Stranger!")
else:
    print("Hello,", name + ". Good to meet you!")
```

```python
# Que.2: Password confirmation program

# Ask the user to enter a password
pass1 = input("Enter a new password: ")

# Ask the user to re-enter the password
pass2 = input("Confirm your password: ")

# Check if both passwords match
if pass1 == pass2:
    print("Password set successfully.")
else:
    print("Error: Passwords do not match.")
```

```python
# Que.3: Password confirmation with length check

pass1 = input("Enter a new password: ")
pass2 = input("Confirm your password: ")

# Check password length
if len(pass1) < 8 or len(pass1) > 12:
    print("Error: Password must be between 8 and 12 characters long.")
elif pass1 != pass2:
    print("Error: Passwords do not match.")
else:
    print("Password set successfully.")
```

```python
# Que.4: Block common (bad) passwords

BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

pass1 = input("Enter a new password: ")
pass2 = input("Confirm your password: ")

if pass1 in BAD_PASSWORDS:
    print("Error: This password is too common.")
elif len(pass1) < 8 or len(pass1) > 12:
    print("Error: Password must be between 8 and 12 characters long.")
elif pass1 != pass2:
    print("Error: Passwords do not match.")
else:
    print("Password set successfully.")
```

```python
# Que.5: Keep asking until a valid password is chosen

BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

while True:
    pass1 = input("Enter a new password: ")
    pass2 = input("Confirm your password: ")

    if pass1 in BAD_PASSWORDS:
        print("Error: This password is too common.")
    elif len(pass1) < 8 or len(pass1) > 12:
        print("Error: Password must be between 8 and 12 characters long.")
    elif pass1 != pass2:
        print("Error: Passwords do not match.")
    else:
        print("Password set successfully.")
        break                 # Exits the loop when password is valid
```

```python
# Que.6: Times Table of seven

for i in range(0, 13):
    print(i, "x 7 =", i * 7)
```

```python
# Que.7: Times table choosed by the user

table = int(input("Enter a number between 0 and 12: "))

# Check if number is within range
if table < 0 or table > 12:
    print("Error: Number must be between 0 and 12.")
else:
    for i in range(0, 13):
        print(i, "x", table, "=", i * table)  #Displays the table of the number entered by the user
```

```python
# Que.8: Reverse times table for negative input

table = int(input("Enter a number: "))

# If input number is negative, print table backwards
if table < 0:
    table = abs(table)  # Convert to positive
    for i in range(12, -1, -1):
        print(i, "x", table, "=", i * table)
else:
    for i in range(0, 13):
        print(i, "x", table, "=", i * table)
```
